### Data wrangling adding WeekYr and WeekNum

```python
def year (series):
    if series <= 8:
        return 2019
    else:
        return 2020

CFA['WeekYr']= CFA['Week'].apply(year)
```

```python
CFA.tail()
```

```python
def wk (series):
    if series <= 2019:
        return 44
    else:
        return -8
    
CFA['WeekF'] = CFA['WeekYr'].apply(wk)
CFA['WeekNum'] = CFA["Week"] + CFA['WeekF']
```

```python
CFA
```

```python
CFA.drop(['WeekF'],axis=1, inplace=True)
CFA.tail()
```

```python
CFA['Productivity'] = CFA['Sales'] / CFA['HoursUsed']
CFA.head()
```

```python
def division (series):
    if series <= 19:
        return 'Before'
    if series >= 25:
        return 'After'
    else:
        return 'During'
    
CFA['Time'] = CFA['Week'].apply(division)
CFA
```
